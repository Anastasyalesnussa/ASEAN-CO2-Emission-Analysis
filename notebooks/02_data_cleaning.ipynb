```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv("co2_emission_asean_raw.csv")

df.head()
```

```python
df.columns = df.columns.str.strip()
```

```python
df = df.rename(columns={
    "Country": "country",
    "Year": "year",
    "Annual CO2 emissions (per capita)": "co2_per_capita"
})
```

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(' ', '_')
              .str.replace('₂', '2')
              .str.replace('__', '_')    
              .str.replace('[()]', '', regex=True)
)

df = df.rename(columns={
    "annual_co2_emissions_per_capita": "co2_per_capita"
})

print(df.columns.tolist())
```

```python
df["year"] = df["year"].astype(int)
df["co2_per_capita"] = pd.to_numeric(df["co2_per_capita"], errors="coerce")
```

```python
df.to_csv("co2_emission_asean_clean.csv", index=False)
print("✅ Clean dataset saved successfully!")
```
